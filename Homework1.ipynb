````python
"""The following code plotting the sales data in a hyperplane improves preformance."""
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.formula.api as sm

#hyperplane
df_3d = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
model = sm.ols(formula='sales ~ TV + radio', data = df_3d)
fit = model.fit()

fig = plt.figure(figsize=(15,15))
axis = fig.add_subplot(111, projection='3d')

x_axis = np.arange(0, 350, 20)                
y_axis = np.arange(0, 60, 4)
x_axis, y_axis = np.meshgrid(x_axis, y_axis)

exog = pd.core.frame.DataFrame({'TV': x_axis.ravel(), 'radio': y_axis.ravel()})
out = fit.predict(exog = exog)
axis.plot_surface(x_axis, y_axis,
                out.reshape(x_axis.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.4)

axis.scatter(df_3d['TV'], df_3d['radio'], df_3d['sales'], c='red', marker='o')

axis.set_xlabel('TV')
axis.set_ylabel('radio')
axis.set_zlabel('sales')

plt.show()
````

## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

````python
""""Learning rate defines the size of steps when descending to reach the minimum. 
Higher learning rates are when larger steps are taken to reach the lowest point, but there's a risk of
overshooting that low point because the slope of the descent is always changing and in taking larger steps-
the final step may go past the lowest point.
Lower learning rates mean that smaller steps are taken. As a result, movement down the negative gradient is more accurate / 
applicable because of frequent recalculation. 
The lower learning rate is more precise but at a cost of time - it reaches the minimum point and converges more slowly.
A learning rate is at a good value is finding a medium between when the faster steps are not too large so there is no risk of 
overshooting, and when the smaller and more precise steps that take up too much time or iterations."""
````

````python
from numpy import *
    # The learning rate defines how fast our model learns (converges).
    # y = mx + b
    # m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(x):
    points = genfromtxt("/Users/vzogh/APMAE4990-/data/data.csv", delimiter=",")
    learning_rate = .0001
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = x
    print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')

if __name__ == '__main__':
    run(100)
````
